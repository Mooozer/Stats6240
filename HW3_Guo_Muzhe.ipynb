Question: Given the above transaction data t1 to t7, assume: minsup = 30%, minconf = 80%, find all the frequent itemset, calculate support. (5 points), generate association rules from the itemset, calculate support and confidence (5 points).

Frequent itemset example:
<br>{Chicken, Clothes, Milk} [sup = 3/7]

Association rules example:
<br>Clothes â†’ Milk, Chicken [sup = 3/7, conf = 3/3]


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
records = [['Beef', 'Chicken', 'Milk'] , 
           ['Beef', 'Cheese'], 
           ['Cheese', 'Boots'], 
           ['Beef', 'Chicken', 'Cheese'],
           ['Beef', 'Chicken', 'Clothes', 'Cheese', 'Milk'], 
           ['Chicken', 'Clothes', 'Milk'],
           ['Chicken', 'Milk', 'Clothes']]
association_rules = apriori(records, min_support=0.3, min_confidence=0.8)
association_results = list(association_rules)
```

### (1). find all the frequent itemset, calculate support.

```python
Frequent_itemset = pd.DataFrame({'itemset':[], 'support': []}) 
for i in range(len(association_results)):
    Frequent_itemset.loc[i] = [set(association_results[i][0]), association_results[i][1]]
Frequent_itemset
```

### (2). generate association rules from the itemset, calculate support and confidence

```python
index = 0 
Association_rules = pd.DataFrame({'Association':[], 'support': [], 'confidence': []}) 
for i in range(len(association_results)):
    for j in range(len(association_results[i][2])): 
        Association = str(list(association_results[i][2][j][0])) +'->' + str(list(association_results[i][2][j][1]))
        support = association_results[i][1]
        confidence = association_results[i][2][j][2]
        Association_rules.loc[index] = [Association, support, confidence]
        index += 1
Association_rules
```
